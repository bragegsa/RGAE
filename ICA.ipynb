```python
from sklearn.decomposition import FastICA

import matplotlib.pyplot as plt
import numpy as np

import scipy.io as sio
import os
```

```python
def load_dataset(dataset_mat, dataset_var):

    data_path = os.path.join(os.getcwd(),'datasets/')

    X = sio.loadmat(os.path.join(data_path, dataset_mat))[dataset_var]

    return X
```

```python
dataset_name = 'abu-airport-1'

X = load_dataset(dataset_name, 'data')

x = np.reshape(X, (X.shape[0]*X.shape[1],X.shape[2]))
print(x.shape)

ica=FastICA(n_components=300, whiten=True)
S_ = ica.fit_transform(x) 

A_=ica.mixing_  
W = np.linalg.pinv(A_) 

B_W=np.sum(np.absolute(W),axis=0)
sortB_W =np.argsort(B_W) 
```
