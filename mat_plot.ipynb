```python
import numpy as np
import scipy.io as sio
import os
import matplotlib.pyplot as plt
from configparser import Interpolation
```

```python
def load_dataset(dataset_mat, dataset_var, map_var):

    data_path = os.path.join(os.getcwd(),'datasets/')

    X = sio.loadmat(os.path.join(data_path, dataset_mat))[dataset_var]
    y = sio.loadmat(os.path.join(data_path, dataset_mat))[map_var]

    return X, y

def load_data(dataset_mat, dataset_var, gt_mat, gt_var, result_mat, result_var):

    data_path = os.path.join(os.getcwd(),'plot')

    X = sio.loadmat(os.path.join(data_path, dataset_mat))[dataset_var]
    y = sio.loadmat(os.path.join(data_path, gt_mat))[gt_var]

    # data_path = os.path.join(os.getcwd(),'results')

    r = sio.loadmat(os.path.join(data_path, result_mat))[result_var]

    return X, y, r

def load_results(dataset_name, file_name):

    data_path = os.path.join(os.getcwd(),'results',dataset_name)

    y = sio.loadmat(os.path.join(data_path, file_name))['y']

    return y

def plot_data(X, y, r):

    plt.figure(figsize=(5,5))
    plt.axis('off')
    plt.imshow(X[:,:,10])

    plt.figure(figsize=(5,5))
    plt.axis('off')
    plt.imshow(y)

    plt.figure(figsize=(5,5))
    plt.axis('off')
    plt.imshow(r)
```
