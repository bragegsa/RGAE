```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear(x, a):
    return x*a

def sig(x):
    return 1/(1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def ReLU(x):
    f = []
    for i in x:
        f.append(max([0,i]))
    return f
```

```python
def plot_act(x, f, func_name):
    plt.figure(figsize=(5,5))
    plt.grid()
    plt.xlabel("x") 
    plt.ylabel(func_name)  
    plt.plot(x, f) 
```

```python
x = np.linspace(-10, 10, 50)   

p = linear(x, 1)
plot_act(x, p, 'Linear, a=1')

p = sig(x)
plot_act(x, p, 'Sigmoid(x)')

p = tanh(x)
plot_act(x, p, 'tanh(x)')

p = ReLU(x)
plot_act(x, p, 'ReLU(x)')
```
